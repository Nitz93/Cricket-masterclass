```python
# to convert into dataframes
import pandas as pd

# extract url response/info
from urllib.request import Request, urlopen

# to work with html content
from bs4 import BeautifulSoup 
```

```python
players_data = []
```

```python
# error case: player name = empty string
i = 0
url = 'https://www.cricbuzz.com/profiles/' + str(i)
req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
text = urlopen(req)
player_id = i

soup = BeautifulSoup(text, "lxml")
# player name
soup.findAll('div', attrs = {'class' : 'cb-col-80'})[0].findAll('h1')[0].get_text()
```

```python
# valid case: player name = non empty string
i = 212
url = 'https://www.cricbuzz.com/profiles/' + str(i)
req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
text = urlopen(req)
player_id = i

soup = BeautifulSoup(text, "lxml")
# player name
soup.findAll('div', attrs = {'class' : 'cb-col-80'})[0].findAll('h1')[0].get_text()
```

```python
# select your range, ex: 1 to 20000
for i in range(0, 30000):
    url = 'https://www.cricbuzz.com/profiles/' + str(i)
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    text = urlopen(req)
    player_id = i
    
    try:
        soup = BeautifulSoup(text, "lxml")
        player = soup.findAll('div', attrs = {'class' : 'cb-col-80'})[0].findAll('h1')[0].get_text()

        if player != '':
            img_url = soup.findAll('div', attrs = {'class' : 'cb-col-20'})[0].findAll('img')[0]["src"]
            ls = soup.findAll('div', attrs = {'class' : 'cb-col-60'})
            born = ls[0].get_text()
            birth_place = ls[1].get_text()
            height = ls[2].get_text()
            role = ls[3].get_text()
            bat_style = ls[4].get_text()
            bowl_style = ls[5].get_text()

            players_json = {'player_id': player_id,
                            'player': player,
                            'img_url': img_url,
                            'born': born,
                            'birth_place': birth_place,
                            'height': height,
                            'role': role,
                            'bat_style': bat_style,
                            'bowl_style': bowl_style}

            players_data.append(players_json)

    #         print(i, player)
        else:
            print('ERROR - ', i)
    except:
        print('Exception - ', i)
```

```python
df = pd.DataFrame(players_data)
```

```python
filename = 'raw_playerprofile_dataset_updated.csv'
df.to_csv(filename, index = False)
```
