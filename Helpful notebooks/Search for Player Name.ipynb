```python
import pandas as pd
import Levenshtein as lev
import warnings
warnings.filterwarnings('ignore')

#to display all rows columns 
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)  
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', -1)
```

```python
df = pd.read_csv('deliveries_updated.csv')
```

```python
# all available batsmen names
df.batsman.unique()
```

```python
# search for a player of your choice - assign your own name
player_name = 'DeVilliers' 
```

```python
# create a new df to find the most matching name
pdf = df['batsman']
pdf = pdf.drop_duplicates(keep='first')
pdf = pd.DataFrame(pdf)
pdf.reset_index(inplace = True, drop = True)

pdf['match_player'] = player_name
pdf['matching'] = pdf['batsman'].apply(lambda x: lev.ratio(x.lower(), player_name.lower()))
```

```python
# top 3 matching names
pdf.sort_values('matching', ascending=False).head(3)
```
